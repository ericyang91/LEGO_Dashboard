```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Import and store the data into a DataFrame
df = pd.read_csv('lego_sets.csv')

df.head()
```

```python
# Check the number of rows and columns in the DataFrame
print(df.shape)
print(f'\nThere are {df.shape[0]} rows and {df.shape[1]} columns in the DataFrame.')
```

```python
# Check the count of non-null values and the data type of each column
df.info()

# Rename the columns
df.rename(columns={'set_id':'Set ID', 
                   'name':'Name', 
                   'year':'Year', 
                   'theme': 'Theme', 
                   'subtheme':'Sub-Theme', 
                   'themeGroup':'Theme Group', 
                   'category':'Category', 
                   'pieces':'Piece Count', 
                   'minifigs': 'Minifig Count', 
                   'agerange_min': 'Age Range (Min)', 
                   'US_retailPrice': 'US Retail Price',
                   'bricksetURL': 'Brick Set URL', 
                   'thumbnailURL':'Thumbnail URL', 
                   'imageURL':'Image URL'}, inplace=True)

print(f'\n\nColumn "{df.columns[7]}" is in "{df[df.columns[7]].dtype}" and needs to be converted to "Integer".')
print(f'Column "{df.columns[8]}" is in "{df[df.columns[8]].dtype}" and needs to be converted to "Integer".')
print(f'Column "{df.columns[9]}" is in "{df[df.columns[9]].dtype}" and needs to be converted to "Integer".')
```

```python
# Take a deeper look at the values in 'Piece Count'
print(df['Piece Count'].value_counts(ascending=False).head(10))

# Now, convert the data type into Integer
df['Piece Count'] = df['Piece Count'].astype('Int64')

print(f"\nData type has been converted to Integer.")
```

```python
# Take a deeper look at the values in 'Minifig Count'
print(df['Minifig Count'].value_counts(ascending=False).head(10))

# Now, convert the data type into Integer
df['Minifig Count'] = df['Minifig Count'].astype('Int64')

print(f"\nData type has been converted to Integer.")
```

```python
# Take a deeper look at the values in 'Age Range (Min)'
print(df['Age Range (Min)'].value_counts(ascending=False).head(10))

# Now, convert the data type into Integer
df['Age Range (Min)'] = df['Age Range (Min)'].astype('Int64')

print(f"\nData type has been converted to Integer.")
```

```python
# Check the change
df.info()

print(f"\n\nData types have been changed.")
```

```python
# Do a deeper dive of the numeric columns
print(df.describe())
print(f'\n\nThere are products where there is no LEGO piece. We need to explore and find out why this is the case.')
```

```python
# We now need to decide how we will handle the <NA> or NaN values
print("'Sub-Theme', 'Theme Group', 'Piece Count', 'Minifig Count', 'Age Range (Min)', 'US Retail Price' have null values.")

df.loc[df['Sub-Theme'].isna()].head(30)
```

```python
missing_subtheme = df[df['Sub-Theme'].isna()]
grouped = missing_subtheme.groupby(['Category', 'Theme Group', 'Theme']).size().reset_index(name='Missing Count')
grouped = grouped.sort_values('Missing Count', ascending=False)

print("\nGrouped Missing Sub-Theme Stats:")
print(grouped.head(30))  # top 30 if many rows
```
